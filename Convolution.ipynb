```
import numpy as np
import imageio as io
from matplotlib import pyplot as plt
```

```
#Načítanie obrázku
image = io.imread("https://raw.githubusercontent.com/PavolGrofcik/INANU/master/Datasets/Image.raw.jpg")
```

```
plt.imshow(image)
```

```
image.shape
```

```
height,width = image.shape[:2]
```

## Sharpen Kernel

```
kernel = np.array([[0,-1,0], [-1, 5, -1], [0, -1, 0]])
```

```
kernel.shape
```

```
#Now we need to create zero matrix with the same size as image
new = np.zeros(image.shape)
```

```
new.shape
```

```
print(image[0,0,0])
```

```
image[0,0].shape
```

```
image[height-1,width-1,0]
```

```
#Convolution for red layer
for x in range(0,height):
    for y in range(0,width):
        
        sum = 0.0
        sum2 = 0.0
        sum3 = 0.0
        for a in range(0,3):
            for b in range(0,3):
                
                if ((x-1 + a < 0) or (y-1 + b < 0)
                   or (x-1 + a >= height) or (y-1+b >= width)):
                    continue
                else:
                    sum += kernel[a][b] * image[x-1+a,y-1+b,0]
                    sum2 += kernel[a][b] * image[x-1+a, y-1+b,1]
                    sum3 += kernel[a][b] * image[x-1 +a, y-1+b,2]
        #print("X %d Y %d SUM: %f" % (x,y,sum))
        if(sum > 255):
            sum = 255
        if (sum2 > 255):
          sum2 = 255
        if (sum3 > 255):
          sum3 = 255
         
        if (sum < 0):
          sum = 0
        if (sum2 < 0):
          sum2 = 0
        if (sum3 < 0):
          sum3 = 0
        
        new[x,y,0] = sum
        new[x,y,1] = sum2
        new[x,y,2] = sum3
        
        if(x == 3000):
          print("Half almost done!")
```

```
plt.imshow(new)
```

```
io.imsave("Convolution.jpg", new)
```
